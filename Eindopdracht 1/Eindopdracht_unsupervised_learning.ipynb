##### Schoon dataset op van europarlementariers (die te weinig spreken) en partijen die onder meerdere namen voorkomen

```python
speakers = speakers[~speakers['political'].isin(['EP/CDA', 'EP/CU/SGP', 'EP/D66', 
                                           'EP/GroenLinks', 'EP/PvdA', 'EP/PvdD',
                                           'EP/SP', 'EP/VVD'])]
speakers = speakers.dropna()
speakers = speakers.replace('Forum voor Democratie','FvD')
speakers = speakers.replace('ChristenUnie','CU')
print(speakers.groupby('political').count())
```

##### voeg de texten en sprekers samen zodat duidelijk wordt welke tekst bij welke partij hoort. Daarnaast vallen in dit proces ook incidentele sprekers zoals formateurs en burgers af, zij spreken niet namens een politieke partij

```python
merged = speaker_items.merge(speakers, left_on='speaker', right_on='id')[['id_y','txt','political']]
merged.groupby('political').count().sort_values('id_y', ascending = True)
```

##### Maak een stem-functie om woorden sterkere betekenissen te geven.

```python
stemmer = DutchStemmer()
analyzer = CountVectorizer().build_analyzer()

def stemmed_words(doc):
    return (stemmer.stem(w) for w in analyzer(doc))
```

## Clustering op basis van tekstinhoud

#### Kunnen we de partijen herkennen in een clustering op basis van de woorden die in een spreekbeurt zitten?

```python
kmText = merged[['txt','political']]
```

```python
Vectorize = CountVectorizer(analyzer = stemmed_words, ngram_range=(1,1), max_df = 0.7)
vecText = Vectorize.fit_transform(kmText.txt)
tfidfVecText = TfidfTransformer().fit_transform(vecText)
```

```python
t = time.time()
n = kmText['political'].nunique()
kmText['predicted'] = KMeans(n_clusters=n, random_state=0, max_iter = 10).fit_predict(tfidfVecText)
print(time.time() - t)
```

```python
dfParty = pd.DataFrame(index = kmText['political'].unique())
for i in range(n):
    col = kmText[kmText['predicted'] == i].groupby('political').count()
    dfParty['cluster' + str(i)] = dfParty.join(col)['txt']
dfParty = dfParty.fillna(0)
dfParty = dfParty.div(dfParty.sum(axis=1), axis=0)
dfParty
```

Wat we hier verwachten te zien is hoge percentages. De getallen geven aan welk percentage van elke partij in een cluster zit. Een hoog percentage zou betekenen dat er een groot deel van de partij de zelfde dingen verkondigd.
<br>
<br>
Wat we terug zien is een mooie verdeling van elke partij. De partijen zijn niet precies gelijk verdeeld over elk cluster, maar ze hebben ook geen grote uitschieters onder elkaar. Misschien spreken de partijen wel grotendeels over de zelfde onderwerpen...

```python
t = time.time()
kmText['predicted'] = KMeans(n_clusters=3, random_state=0, max_iter = 10).fit_predict(tfidfVecText)
print(time.time() - t)
```

```python
dfParty = pd.DataFrame(index = kmText['political'].unique())
for i in range(3):
    col = kmText[kmText['predicted'] == i].groupby('political').count()
    dfParty['cluster' + str(i)] = dfParty.join(col)['txt']
dfParty = dfParty.fillna(0)
dfParty = dfParty.div(dfParty.sum(axis=1), axis=0)
dfParty
```

Wanneer we clusteren in drieën, met de verwachting dat er een verdeling links-midden-rechts te zien is komt dit ook niet naar voren. Het algoritme deelt alle partijen in tweeën en stopt een aantal speeches in de laatste cluster. Ook hier is af te vragen hoe veel de inhoud van spreekbeurten per partij verschilt...

## Clustering op basis van spreekfrequentie en lengte

##### Kunnen we clusteren op basis van hoe vaak een speker spreekt en hoe lang zij/hij spreekt?

```python
merged['length'] = merged['txt'].apply(len)
kmSpeakers = pd.DataFrame()
kmSpeakers['nTurns'] = merged.groupby('id_y').count()['txt']
kmSpeakers['mTurnLength'] = merged.groupby('id_y').sum()['length'] / kmSpeakers['nTurns']
kmSpeakers = kmSpeakers.merge(speakers,left_on = 'id_y',right_on = 'id')[['id','nTurns','mTurnLength','political','prefix']]
kmSpeakers.head()
```

#### Bepaalt de functie (prefix) van de spreker zijn of haar spreekfrequentie en spreeklengte?

```python
n = kmSpeakers['prefix'].nunique()
kmSpeakers['predicted'] = KMeans(n_clusters=n, random_state=0).fit_predict(kmSpeakers[['nTurns','mTurnLength']])
```

```python
dfFunctions = pd.DataFrame(index = kmSpeakers['prefix'].unique())
for i in range(n):
    col = kmSpeakers[kmSpeakers['predicted'] == i].groupby('prefix').count()
    dfFunctions['cluster' + str(i)] = dfFunctions.join(col)['id']
dfFunctions = dfFunctions.fillna(0)
dfFunctions = dfFunctions.div(dfFunctions.sum(axis=1), axis=0)

dfFunctions
```

```python
plt.scatter(kmSpeakers['nTurns'],kmSpeakers['mTurnLength'], c = kmSpeakers['predicted'])
plt.xlabel('Spreekfrequentie')
plt.ylabel('Gemiddelde spreeklengte (letters)')
plt.show()
```

Die twee uitschieters zijn mevr. Ploumen (in functie kamerlid) en dhr. Hiddema, zij hebben allebei maar twee keer gesproken en waren toen nogal lang van stof. Zie hier onder de resultaten zonder deze sprekers.

```python
kmSpeakers = kmSpeakers[kmSpeakers['id'] != 136]
kmSpeakers = kmSpeakers[kmSpeakers['id'] != 123]
```

```python
n = kmSpeakers['prefix'].nunique()
kmSpeakers['predicted'] = KMeans(n_clusters=n, random_state=0).fit_predict(kmSpeakers[['nTurns','mTurnLength']])
```

```python
dfFunctions = pd.DataFrame(index = kmSpeakers['prefix'].unique())
for i in range(n):
    col = kmSpeakers[kmSpeakers['predicted'] == i].groupby('prefix').count()
    dfFunctions['cluster' + str(i)] = dfFunctions.join(col)['id']
dfFunctions = dfFunctions.fillna(0)
dfFunctions = dfFunctions.div(dfFunctions.sum(axis=1), axis=0)

dfFunctions
```

```python
plt.scatter(kmSpeakers['nTurns'],kmSpeakers['mTurnLength'], c = kmSpeakers['predicted'])
plt.xlabel('Spreekfrequentie')
plt.ylabel('Gemiddelde spreeklengte (letters)')
plt.show()
```

Er zijn vier groepen te zien: sprekers die vaak weinig zeggen, sprekers die vaak vaak zeggen, sprekers die regelmatig een normale lengte praten en sprekers die veel spreken. Er is vooral te zien dat er een cluster is met regeringsleden. Over de rest van de clusters zijn de groepen redelijk verdeeld.
