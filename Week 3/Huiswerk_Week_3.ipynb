```python
%matplotlib inline
import numpy as np
import scipy.stats as st
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.svm import SVC
```

```python
# maak een dataset aan die makkelijk te scheiden is
# een label met alleen maar waarden tussen 0 en 53 en een met waarden tussen 47 en 100 (zodat er overlap is)

n = 500

Ax1 = np.random.randint(0,53,int(n/2))
Bx1 = np.random.randint(47,100,int(n/2))
x1  = np.append(Ax1,Bx1)

Ax2 = np.random.randint(0,53,int(n/2))
Bx2 = np.random.randint(47,100,int(n/2))
x2  = np.append(Ax2,Bx2)

y1 = [0] * int(n/2)
y2 = [1] * int(n/2)
y = np.append(y1,y2)

dataSimple = pd.DataFrame({
    'X1': x1,
    'X2': x2,
    'Y':  y
})
```

```python
# maak een scatterplot van de datapunten inclusief x=y lijn

plt.axhline(0, color='gray', zorder=1)
plt.axvline(0, color='gray', zorder=1)
plt.scatter(dataSimple.X1[dataSimple.Y == 1], dataSimple.X2[dataSimple.Y == 1], marker='o', s=100, zorder=2, color='b')
plt.scatter(dataSimple.X1[dataSimple.Y == 0], dataSimple.X2[dataSimple.Y == 0], marker='s', s=100, zorder=2, color='orange')
plt.plot([0, 100], [0, 100], color='r')
plt.axis('equal')
plt.show()
```

```python
# fit een SVM model met een lineaire kernel

clf = SVC(C=1, gamma='scale', kernel='linear')
clf.fit(dataSimple[['X1', 'X2']], dataSimple['Y'])
```

```python
# bereid de scheidingslijn voor voor visualisatie

xx, yy = np.meshgrid(np.linspace(-1, 101, 100), np.linspace(-1, 101, 100))
yhat = clf.predict(np.c_[xx.ravel(), yy.ravel()])
yhat = yhat.reshape(xx.shape)
```

```python
# plot alle datapunten met scheidingslijn

plt.figure(figsize=(10, 10))
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.axhline(0, color='gray', zorder=1)
plt.axvline(0, color='gray', zorder=1)

plt.pcolormesh(xx, yy, yhat, cmap='Pastel2', zorder=0)
plt.scatter(dataSimple.X1[dataSimple.Y == 1], dataSimple.X2[dataSimple.Y == 1], marker='o', s=100, zorder=2, color='b')
plt.scatter(dataSimple.X1[dataSimple.Y == 0], dataSimple.X2[dataSimple.Y == 0], marker='s', s=100, zorder=2, color='orange')
plt.plot([0, 100], [0, 100], color='r')
plt.axis('equal')

plt.show()
```

```python
# maak een dataset aan die niet makkelijk te scheiden is
# een label met alleen maar waarden binnen een cirkel van straal 70 vallen en punten die buiten deze cirkel vallen

n = 500

x1 = np.random.randint(0,100,n)
x2 = np.random.randint(0,100,n)
y  = []

for i in range(len(x1)):
    if x1[i]**2 + x2[i]**2 < 70**2:
        y.append(0)
    else:
        y.append(1)

dataMixed = pd.DataFrame({
    'X1': x1,
    'X2': x2,
    'Y':  y
})
```

```python
# maak een scatterplot van de datapunten inclusief x=y lijn

plt.axhline(0, color='gray', zorder=1)
plt.axvline(0, color='gray', zorder=1)
plt.scatter(dataMixed.X1[dataMixed.Y == 1], dataMixed.X2[dataMixed.Y == 1], marker='o', s=100, zorder=2, color='b')
plt.scatter(dataMixed.X1[dataMixed.Y == 0], dataMixed.X2[dataMixed.Y == 0], marker='s', s=100, zorder=2, color='orange')
plt.plot([0, 100], [0, 100], color='r')
plt.axis('equal')
plt.show()
```

```python
# fit een SVM model met een rbf kernel

clf = SVC(C=1, gamma='scale', kernel='rbf')
clf.fit(dataMixed[['X1', 'X2']], dataMixed['Y'])
```

```python
# bereid de scheidingslijn voor voor visualisatie

xx, yy = np.meshgrid(np.linspace(-1, 101, 100), np.linspace(-1, 101, 100))
yhat = clf.predict(np.c_[xx.ravel(), yy.ravel()])
yhat = yhat.reshape(xx.shape)
```

```python
# plot alle datapunten met scheidingslijn

plt.figure(figsize=(10, 10))
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.axhline(0, color='gray', zorder=1)
plt.axvline(0, color='gray', zorder=1)

plt.pcolormesh(xx, yy, yhat, cmap='Pastel2', zorder=0)
plt.scatter(dataMixed.X1[dataMixed.Y == 1], dataMixed.X2[dataMixed.Y == 1], marker='o', s=100, zorder=2, color='b')
plt.scatter(dataMixed.X1[dataMixed.Y == 0], dataMixed.X2[dataMixed.Y == 0], marker='s', s=100, zorder=2, color='orange')
plt.plot([0, 100], [0, 100], color='r')
plt.axis('equal')

plt.show()
```
